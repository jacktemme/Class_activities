```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Load in file
# Store filepath in a variable
movie_file = Path("Resources/movie_scores.csv")
```

```python
# Read and display the CSV with Pandas
movie_df = pd.read_csv(movie_file, encoding="utf-8")
movie_df.head()
```

```python
# List all the columns in the table

movie_df.columns
```

```python
# We only want IMDb data, so create a new table that takes the Film and all the columns relating to IMDB
imdb_df = movie_df[['IMDB','IMDB_norm','IMDB_norm_round','IMDB_user_vote_count']]
imdb_df.head()
```

```python
good_movie_df = movie_df.loc[movie_df["IMDB"] > 7, ["FILM", "IMDB"]]
good_movie_df.sort_values(by=["IMDB"], ascending = False)
```

```python
# Find less popular movies--i.e., those with fewer than 20K votes
unpop_movies_df = movie_df.loc[movie_df["IMDB_user_vote_count"]<20000, ["FILM","IMDB_user_vote_count"]]
unpop_movies_df.sort_values(by = ["IMDB_user_vote_count"])
```

```python
# Finally, export this file to a spread so we can keep track of out new future watch list without the index
unpop_movies_df.to_csv("Resources/movies.csv", index = True)
```

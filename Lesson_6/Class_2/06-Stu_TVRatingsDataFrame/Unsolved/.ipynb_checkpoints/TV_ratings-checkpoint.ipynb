```python
# Dependencies
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
# list of TV show titles to query
tv_shows = ["The Walking Dead","Breaking Bad", "Spongebob Squarepants", "Jujutsu Kaisen", "Vinland Saga", "Attack on Titan"]

# TV Maze show search base URL
base_url = "http://api.tvmaze.com/search/shows?q="

# set up lists to hold response data for name and rating
name = []
rating = []
# loop through TV show titles, make requests and parse
for show in tv_shows:
    query_url = (f'{base_url}{show}')
    response = requests.get(query_url).json()
    name.append(response[0]['show']['name'])
    rating.append(response[0]['show']['rating']['average'])
```

```python
# create DataFrame
tv_data = {"Title": name,"Rating":rating}
tv_df = pd.DataFrame(tv_data)
tv_df
```

```python
# Plot the data
plt.bar(name,rating, color = "green")
plt.xlabel("TV Shows")
plt.ylabel("Rating")
plt.xticks(rotation = 45, ha = "right", )
plt.figure(figsize = (10,5))
plt.tight_layout()
plt.show()
```

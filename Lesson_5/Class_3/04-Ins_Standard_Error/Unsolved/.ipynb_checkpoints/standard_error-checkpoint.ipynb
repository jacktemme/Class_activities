```python
# Dependencies
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import sem

# Set the seed so our data is reproducible
random.seed(42)
```

```python
# Sample versus population example fuel economy
fuel_economy = pd.read_csv('../Resources/2019_fuel_economy.csv')

# First overview the data set - how many factors, etc.
print(fuel_economy.head())
```

```python
# Calculate the summary statistics and plot the histogram of the entire population data
print(np.mean(round(fuel_economy["Combined_MPG"],2)))
print(np.std(round(fuel_economy["Combined_MPG"],2)))
plt.hist(fuel_economy["Combined_MPG"])
plt.xlabel("Miles Per Gallon")
plt.ylabel("Count")
plt.show()
```

```python
# Calculate the summary statistics and plot the histogram of the sample data using iloc
partial = fuel_economy.iloc[range(700,900)]
print(np.mean(round(partial["Combined_MPG"],2)))
print(np.std(round(partial["Combined_MPG"],2)))
plt.hist(partial["Combined_MPG"])
plt.xlabel("Miles Per Gallon")
plt.ylabel("Count")
plt.show()
```

```python
# Calculate the summary statistics and plot the histogram of the sample data using random sampling
rand = fuel_economy.sample(42)
print(np.mean(round(rand["Combined_MPG"],2)))
print(np.std(round(rand["Combined_MPG"],2)))
plt.hist(rand["Combined_MPG"])
plt.xlabel("Miles Per Gallon")
plt.ylabel("Count")
plt.show()
```

```python
# Generate a new 30 vehicle sample and calculate the SEM of the sample
mpg = fuel_economy["Combined_MPG"]
sample = mpg.sample(30) 

standard_error= sem(sample)
print(standard_error)
```

```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
# Read in the california housing data set
california_data = pd.read_csv('../Resources/California_Housing.csv')
california_data.head()
```

```python
# Get the information on the DataFrame
california_data.info()
```

```python
# Determine which measure of central tendency is most appropriate to describe the Population
population = california_data['Population']
median = np.median(population)
mean = np.mean(population)
mode = st.mode(population)
print(median)
print(mean)
print(mode)
```

```python
california_data["Population"].plot(kind = "hist")
plt.show()
```

```python
# Determine if the house age in California is normally distributed using a small and large sample size. 

small_sample = st.normaltest(california_data["HouseAge"].sample(n = 100))
large_sample = st.normaltest(california_data["HouseAge"].sample(n = 1000))
print(small_sample)
print(large_sample)
plt.hist(california_data["HouseAge"])
```

```python
# Determine if there are any potential outliers in the average occupancy in California
occupancy = california_data["AveOccup"]
quartiles = occupancy.quantile([.25,.5,.75])
lowerq = quartiles[.25]
upperq = quartiles[.75]

iqr = upperq-lowerq

upper_bound = (iqr * 1.5) + upperq
lower_bound = (iqr * 1.5) - lowerq

california_outliers = california_data.loc[occupancy > upper_bound]
california_outliers
```

```python
# With the potential outliers, what is the lowest and highest median income (in $1000s) observed?
min_income = np.min(california_outliers["MedInc"])
max_income = np.max(california_outliers["MedInc"])
print(min_income)
print(max_income)
```

```python
# Bonus - plot the latitude and longitude of the California housing data using Matplotlib, color the data points using the median income of the block.

scatter = plt.scatter(california_data["Longitude"],california_data["Latitude"], c = california_data["MedInc"])
plt.colorbar(scatter, label = "Median Income")
plt.show()
```

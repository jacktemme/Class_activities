```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
combined_travel_df = pd.read_csv("Resources/combined_travel.csv")
combined_travel_df.head()
```

```python
# set up data to use for plot
country_df = combined_travel_df.set_index("COUNTRY OF NATIONALITY")
companion_df = country_df[["2016 Alone","2017 Alone","2018 Alone"]]
companion_df = companion_df.iloc[[0,2,18]]
companion_df 
```

```python
x_axis = [2017,2018,2019]
```

```python
plt.plot(x_axis, companion_df.iloc[0].values, color = "green", label = "Singapore")
plt.plot(x_axis, companion_df.iloc[1].values, color = "red", label = "Indonesia" )
plt.plot(x_axis, companion_df.iloc[2].values, color = "blue", label = "USA")
plt.xticks(x_axis)
plt.legend(loc = "upper right")
plt.show()
```

```python
plt.title("Traveling Alone Country Comparison")
plt.ylabel ("% of Travelers")
plt.xlabel ("Years")
```

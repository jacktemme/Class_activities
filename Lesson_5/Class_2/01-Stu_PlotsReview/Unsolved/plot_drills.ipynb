```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA SET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(0, len(gyms))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

plt.title("NYC Gym Popularity")
plt.xlabel("Gym Name")
plt.ylabel("Number of Members")

# Finish the plot
plt.bar(x_axis, members, color = "red", align = "center")
plt.xticks(tick_locations,gyms)
plt.show()
```

```python
# DATA SET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.title("Sin from 0 to 2$\pi$")
plt.xlabel("Real Numbers from 0 to 2$\pi$")
plt.ylabel("sin(x)")

# Finish the plot
plt.hlines(0,0,x_lim,alpha=.50)
plt.xlim(0,x_lim)
plt.ylim(-1.3,1.3)
plt.plot(x_axis,sin, color = "orange", marker = "^")
plt.show()
```

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
car_data = pd.read_csv('../Resources/mpg.csv')
car_data.head()
```

```python
# Remove the rows with missing values in horsepower
car_filter_data = car_data.loc[car_data["horsepower"] != "?"]
```

```python
# Set the 'car name' as our index
car_filter_data.set_index(["car name"])

# Remove the 'origin' column

car_filter_data.head()
```

```python
# Convert the "horsepower" column to numeric so the data can be used
car_filter_data["horsepower"] = car_filter_data["horsepower"].astype('float64')
car_filter_data.dtypes
```

```python
# Create a scatter plot which compares MPG to horsepower
car_filter_data.plot(kind = "scatter", x = "horsepower", y = "mpg", title = "MPG versus Horsepower", grid = True)
plt.show()
```

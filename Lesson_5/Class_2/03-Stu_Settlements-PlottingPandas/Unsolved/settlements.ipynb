```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
union_df = pd.read_csv("Resources/union_settlements_1995.csv")
union_df.head()
```

```python
# List columns
union_df.columns
```

```python
# Get total settlements by union
union_df1 = union_df["UNION"].value_counts()
union_df1
```

```python
# Configure plot, figsize, title, and axis labels
figure = union_df1.plot(kind = "bar", color = "orange", alpha = .75, title = "1995 Union Settlements",
                xlabel = "Settlement", ylabel = "Total Settlements" )

# Configure x-tick rotation
xtick_label = union_df1.index
figure.set_xticklabels(xtick_label, rotation = 45, rotation_mode = "anchor", ha = "right", wrap = True)
plt.tight_layout()
plt.show()
```

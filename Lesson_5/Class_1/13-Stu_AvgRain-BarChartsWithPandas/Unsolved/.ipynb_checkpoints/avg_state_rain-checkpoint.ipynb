```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Load in csv
csvfile = Path("../Resources/avg_rain_state.csv")
rain_df = pd.read_csv(csvfile)
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(0,len(rain_df))
tick_location = [value + 0.6 for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize = (15,5))
plt.bar(x_axis,rain_df["Inches"], color = "red", alpha = 0.6, align = "edge")
plt.xticks(tick_location, rain_df["State"], rotation = "vertical")
plt.show()
```

```python
# Set x and y limits
plt.xlim(-.3, len(x_axis))
plt.ylim(0,max(rain_df["Inches"]))
```

```python
# Set a Title and labels
plt.title("Average Rainfall Per State")
plt.xlabel("State")
plt.ylabel("Inches")
```

```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Load in csv
csvfile = Path("../Resources/avg_rain_state.csv")
rain_df = pd.read_csv(csvfile)
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(0,len(rain_df),1)
tick_location = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize = (12,8))
plt.bar(x_axis,rain_df["Inches"], color = "skyblue", alpha = 0.75)
plt.xticks(tick_location, labels = rain_df["State"], rotation = "vertical")
plt.show()
```

```python
# Set x and y limits
plt.ylim(0,65)
plt.xlim(-1, 52)
```

```python
# Set a Title and labels
plt.title("Yearly Rainfall for each State")
plt.xlabel("States")
plt.ylabel("Rainfall (Inches)")
```

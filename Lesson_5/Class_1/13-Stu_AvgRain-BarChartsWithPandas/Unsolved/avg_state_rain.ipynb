```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Load in csv
csvfile = Path("../Resources/avg_rain_state.csv")
rain_df = pd.read_csv(csvfile)
rain_df.head()
```

```python
# Set x axis and tick locations
x_axis = np.arange(0,len(rain_df["State"]))
tick_location = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize = (12,12))
plt.bar(x_axis,rain_df["Inches"], color = "Red")
plt.xticks(tick_location,rain_df["State"], rotation = "vertical")
plt.show()
```

```python
# Set x and y limits
```

```python
# Set a Title and labels
plt.title("Rainfall Per State")
plt.xlabel("State")
plt.ylabel("Rainfall (Inches)")
```

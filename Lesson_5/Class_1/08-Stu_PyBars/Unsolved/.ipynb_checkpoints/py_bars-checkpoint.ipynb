```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"]
cars_in_cities = [214.7, 564.4, 416.5, 466.7, 350.6]
x_axis = np.arange(len(cars_in_cities))
```

```python
# Create a bar chart based upon the above data
```

```python
plt.bar(x_axis, cars_in_cities)
```

```python
# Create the ticks for our bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations,["San Francisco", "Omaha", "New Orleans", "Cincinnati", "Pittsburgh"] )
plt.show()
```

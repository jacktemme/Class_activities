```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set x axis to numerical value for month
x_axis = np.arange(1,13,1)
x_axis
```

```python
# Avearge weather temp
points_F = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
# Convert to Celsius C = (F-32) * 0.56
celsius = [(y -32)*.56 for y in points_F]
celsius
```

```python
plt.clf()
```

```python
# Create a handle for each plot
plt.plot(x_axis, points_F, marker = "+", color = "blue", label = "Fahrenheit")
plt.plot(x_axis,celsius, marker = "s", color = "red", label = "Celsius")
plt.show()
```

```python
# Set our legend to where the chart thinks is best
plt.legend(loc="upper left")
```

```python
# Create labels for the X and Y axis
plt.xlabel("Months of the Year")
plt.ylabel("Temperature")
```
